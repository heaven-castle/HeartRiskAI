```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cell 2: Load Dataset
df = pd.read_csv('../data/heart_attack_prediction_indonesia.csv')
df.head()

# Cell 3: Info Dataset
df.info()

# Cell 4: Cek Missing Values
df.isnull().sum()

# Cell 5: Statistik Deskriptif
df.describe()
```

```python
sns.countplot(data=df, x='heart_attack')
plt.title("Distribusi Kelas Target (Heart Attack)")
plt.xlabel("Heart Attack (0=Tidak, 1=Ya)")
plt.ylabel("Jumlah")
plt.show()
```

```python
# Cell 7: Korelasi Antarfitur Numerik (tanpa kolom kategorikal)
numeric_df = df.select_dtypes(include=['int64', 'float64'])  # hanya kolom numerik
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriks Korelasi (Kolom Numerik Saja)")
plt.show()
```

```python
numerik_cols = ['age', 'cholesterol_level', 'blood_pressure_systolic', 'blood_pressure_diastolic', 'sleep_hours']
for col in numerik_cols:
    sns.boxplot(data=df, x='heart_attack', y=col)
    plt.title(f'{col} vs Heart Attack')
    plt.show()
```

```python
import pandas as pd

# Cek apakah kolom sudah ada
assert 'blood_pressure_systolic' in df.columns, "Kolom 'blood_pressure_systolic' tidak ditemukan!"

# Kategorisasi tekanan darah sistolik
df['bp_category'] = pd.cut(
    df['blood_pressure_systolic'],
    bins=[0, 120, 140, 180, float('inf')],
    labels=["Normal", "Elevated", "High", "Very High"]
)

# One-hot encoding kategori tekanan darah
bp_dummies = pd.get_dummies(df['bp_category'], prefix='bp')

# Gabungkan ke dataset
df = pd.concat([df, bp_dummies], axis=1)

# (Opsional) Hapus kolom kategori asli jika tidak dibutuhkan
# df.drop(columns=['bp_category'], inplace=True)

# Tampilkan beberapa baris untuk verifikasi
df[['blood_pressure_systolic', 'bp_category'] + list(bp_dummies.columns)].head()
```

```python
cat_cols = ['smoking_status', 'physical_activity']
for col in cat_cols:
    sns.countplot(data=df, x=col, hue='heart_attack')
    plt.title(f'Distribusi {col} berdasarkan Heart Attack')
    plt.xticks(rotation=20)
    plt.show()
```
